```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
train = pd.read_csv("PGATOUR_data2.csv")
```

```python
train.dropna(inplace=True)
```

```python
train
```

```python
train.columns
```

```python
X = train[['EVENTS_PLAYED', 'POINTS','ROUNDS_PLAYED','AVG_Driving_DISTANCE','AVG_CLUB_HEAD_SPEED','AVG_BALL_SPEED','AVG_LAUNCH_ANGLE',
          'AVG_SPIN_RATE','AVG_HANG_TIME','AVG_CARRY_DISTANCE','TOTAL_STROKES','TOTAL_ROUNDS']]
```

```python
y = train['AVG_SCORE']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor 
```

```python
rm = RandomForestRegressor(random_state=0)
```

```python
rm.fit(X_train, y_train)
```

```python
pred = rm.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, pred)
```
